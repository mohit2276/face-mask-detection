```python
import numpy as np
import cv2
```

```python
with_mask=np.load('with_mask.npy')
wihtout_mask=np.load('without_mask.npy')
```

```python
with_mask.shape
```

```python
wihtout_mask.shape
```

```python
#convert 4 dimensional data into 2 dimensional data
with_mask= with_mask.reshape(200,50*50*3)
wihtout_mask=wihtout_mask.reshape(200,50*50*3)

```

```python
with_mask.shape
```

```python
wihtout_mask.shape
```

```python
# r_ is used  to concanate the rows in numpy
#it add 200 rows of both
X=np.r_[with_mask, wihtout_mask]
```

```python
X.shape
```

```python
# now we need to label the mask with mask with  and wihtout mask at

labels=np.zeros(X.shape[0])
```

```python
# data at 200 is withmask and after 200 it is without mask at 1.0

labels[200:]=1.0
```

```python
# if the ouptut is  0 then it wear the mask
# if the output is 1 then it is not wear the mask
names={0:'MASK',1:'NO MASK'}
```

```python

#svm-support vector machine
#svc-support vector classification
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
#accuracy of the algorithm
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,labels,test_size=0.20)

```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#dimensionality reduction 
# used to reduce the dimensions beacuse the data is high it take more time otherwise
#there is a packge decomposition for reduce the dimension of data
#principal component analsyis  -PCA


from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=3)
x_train=pca.fit_transform(x_train)
```

```python
x_train[0]
```

```python
# eigen values and eigen vectors are basic principle used for dimensionally reduction

x_train.shape
```

```python
svm= SVC()
svm.fit(x_train,y_train)
```

```python
x_test=pca.transform(x_test)
y_pred=svm.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
haar_data=cv2.CascadeClassifier('data.xml')
capture=cv2.VideoCapture(0)
data=[]
font = cv2.FONT_HERSHEY_COMPLEX
while True:
    flag,img = capture.read()
    if flag:
        faces=haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img,(x,y),(x+w, y+h), (255,0,255), 4)
            # after detecting face it remove our face with the help of slicing

            face=img[y:y+h, x:x+w, : ]
            # because face has different dimensions so we have to resize it to all size same
             
            face=cv2.resize(face,(50,50))
            face=face.reshape(1,-1)
            face=pca.transform(face)
            pred=svm.predict(face)
            n=names[int(pred)]
            # on showing text on real time
            cv2.putText(img,n,(x,y),font , 1,(244,250,250),2)
            print(n)

           



        cv2.imshow('result',img)
    # 27 is the -ASCII of escape 
        if cv2.waitKey(2) == 27:
            break


#for opening the cam we use this command        
capture.release()
cv2.destroyAllWindows()

```
